```python
import requests
import random
from bs4 import BeautifulSoup

def google_search(query):
    try:
        search_url = f"https://www.google.co.uk/search?q={query}&num=1"
        response = requests.get(search_url)
        soup = BeautifulSoup(response.text, "html.parser")
        links = []
        for h3 in soup.find_all('h3'):
            a_tag = h3.find_parent('a')
            
            if a_tag and 'href' in a_tag.attrs:
                link = a_tag['href']
                link = link.split('/url?q=')[1]
                links.append(link)
        return links[0]
    except Exception as e:
        print(e)
        return ''



```

```python
import pandas as pd

df = pd.read_excel('Updated_list.xlsx')

df
```

```python
df = df.astype(str)
```

```python
import time
for index,row in df.iterrows():
    if row['Linkedin Profile'] == 'nan':
        name = row['Name']
        company = row['Company'] if row['Company']!='Genral' else ''
        search_text = f'{name} linkedin profile UK {company}'

        link = google_search(search_text)

        df.at[index,'Linkedin Profile'] = link
        print(f'Done for {name} and link is {link}')
```

```python
df.to_excel('Updated_list.xlsx',index=False)
```

```python
import openpyxl
from openpyxl.utils import get_column_letter

def convert_urls_to_hyperlinks(excel_file, sheet_name, url_column):
    workbook = openpyxl.load_workbook(excel_file)
    sheet = workbook[sheet_name]

    col_letter = get_column_letter(url_column)

    for row in range(2, sheet.max_row + 1): 
        cell = sheet[f"{col_letter}{row}"]
        if isinstance(cell.value, str) and cell.value.startswith("http"):
            sheet[f"{col_letter}{row}"].hyperlink = cell.value
            sheet[f"{col_letter}{row}"].value = cell.value
            sheet[f"{col_letter}{row}"].style = "Hyperlink"

    workbook.save(excel_file)

excel_file = 'Updated_list.xlsx'
sheet_name = "Sheet1"
url_column = 3  

convert_urls_to_hyperlinks(excel_file, sheet_name, url_column)
```
